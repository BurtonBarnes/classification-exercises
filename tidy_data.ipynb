```python
#1.
import pandas as pd
import numpy as np
from env import host, user, password

url = f'mysql+pymysql://{user}:{password}@{host}/tidy_data'
```

```python
attendance = pd.read_sql('SELECT * FROM attendance', url)
attendance
```

```python
attendance = attendance.melt(id_vars=['Unnamed: 0'], var_name='date', value_name ='present')
attendance
```

```python
atten_rep=attendance.replace({'A':0, 'P':1, 'T':0.9, 'H':0.5})
atten_rep.head()
```

```python
atten_ren=atten_rep.rename(columns = {'Unnamed: 0': 'student'})
atten_ren
```

```python
atten_ren.sort_values(by='student')
```

```python
atten_ren.groupby('student').present.agg(['mean'])
```

```python
#3.a.
cake_recipes = pd.read_sql('SELECT * FROM cake_recipes', url)
cake_recipes
```

```python
#3.b.
cake = cake_recipes.melt(id_vars=['recipe:position'], var_name='oven_temp', value_name='response')
cake
```

```python
cake = cake.rename(columns = {'recipe:position':'recipe_position'})
cake
```

```python
cake_new = cake.recipe_position.str.split(':', expand = True)
cake_new.columns = ['recipe', 'position']
cake_new
```

```python
better_cake = pd.concat([cake,cake_new], axis=1).drop(columns='recipe_position')
better_cake.head()
```
